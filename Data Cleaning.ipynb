```python
import pandas as pd
holiday_filepath = 'https://www.data.gouv.fr/fr/datasets/r/6637991e-c4d8-4cd6-854e-ce33c5ab49d5'
consumption_filepath = 'consommation-quotidienne-brute-regionale.csv'
population_filepath = "POPULATION_MUNICIPALE_REGION_FRANCE.xlsx"
region_depts_filepath = "https://www.data.gouv.fr/fr/datasets/r/987227fb-dcb2-429e-96af-8979f97c9c84"
FRANCE_50_25_filepaths = ["https://www.data.gouv.fr/fr/datasets/r/c79aaafe-8017-4d2b-8884-57b5391da5bc", #1
                          "https://www.data.gouv.fr/fr/datasets/r/98852d18-6580-47d4-9442-0bb3184896c8", #1
                          "https://www.data.gouv.fr/fr/datasets/r/e8cafb02-248c-41c1-a409-8eb5aaf4a102", #2
                          "https://www.data.gouv.fr/fr/datasets/r/720841f1-45c8-4246-8f83-7ac07d0f5cbb", #2
                          "https://www.data.gouv.fr/fr/datasets/r/777fe96d-987e-4e31-ad67-3a8991849263", #3
                          "https://www.data.gouv.fr/fr/datasets/r/c94ced69-cf96-44f6-8d3d-7e64a156a2ab", #3
                          "https://www.data.gouv.fr/fr/datasets/r/eee4b43d-0cc0-4242-8929-10ee9cc67596", #4
                          "https://www.data.gouv.fr/fr/datasets/r/52355223-df85-49a9-9031-14b8a604c28c", #4
                          "https://www.data.gouv.fr/fr/datasets/r/edde6d8d-e026-4dfa-94aa-2a1d5c3c5eac", #5
                          "https://www.data.gouv.fr/fr/datasets/r/27bfba71-f9d2-4cbb-a784-7a893d933485", #5
                          "https://www.data.gouv.fr/fr/datasets/r/ce6518fa-4831-4b31-91eb-7f41c354ef58", #6
                          "https://www.data.gouv.fr/fr/datasets/r/6d848c25-210b-40a4-b3c2-a4084bf660a7", #6
                          "https://www.data.gouv.fr/fr/datasets/r/7f708a21-f6f1-46a7-bee2-a331a579e628", #7
                          "https://www.data.gouv.fr/fr/datasets/r/71391f93-5e3e-42c7-b137-6e0815f80b33", #7
                          "https://www.data.gouv.fr/fr/datasets/r/2ff1a5db-00a3-4f1b-b74e-f898bbb8cab6", #8
                          "https://www.data.gouv.fr/fr/datasets/r/e500252a-f46c-4222-902f-d0c25b309a1d", #8
                          "https://www.data.gouv.fr/fr/datasets/r/e98c097b-a150-4737-9805-c7fb6e71eeec", #9
                          "https://www.data.gouv.fr/fr/datasets/r/5f206327-f167-4d37-9497-136d38f3cf9c", #9
                          "https://www.data.gouv.fr/fr/datasets/r/48186700-71f2-4153-b0b5-b4932b3a670e", #10
                          "https://www.data.gouv.fr/fr/datasets/r/4160a18b-9eb6-4465-803b-b52a1cd000d0", #10
                          "https://www.data.gouv.fr/fr/datasets/r/f5e1ccf2-1ea5-457e-b4fb-7533ae7f9cfb", #11
                          "https://www.data.gouv.fr/fr/datasets/r/2a3f77ba-2e7c-4b73-a754-34a203e690db", #11
                          "https://www.data.gouv.fr/fr/datasets/r/48c22288-7cfc-4f9b-8cd6-081ef314165d", #12
                          "https://www.data.gouv.fr/fr/datasets/r/5e54853f-ec23-4a1c-a5bc-f4fe28c54651", #12
                          "https://www.data.gouv.fr/fr/datasets/r/5669e5be-8c67-4fbf-aeae-08cbc2369dd4", #13
                          "https://www.data.gouv.fr/fr/datasets/r/eb4d0600-e90b-4517-a429-599ed13dbae0", #13
                          "https://www.data.gouv.fr/fr/datasets/r/5feeb9de-d9d3-4ab6-85b4-46bc09ac285a", #14
                          "https://www.data.gouv.fr/fr/datasets/r/42b91ef2-4a43-4193-8a31-0393604c9f07", #14
                          "https://www.data.gouv.fr/fr/datasets/r/587f0b40-ffbb-4a90-83a5-b3b974757bba", #15
                          "https://www.data.gouv.fr/fr/datasets/r/d0b6821d-ab1a-442d-941e-c273e9dd8c8f", #15
                          "https://www.data.gouv.fr/fr/datasets/r/b8c9b479-e535-457d-9192-094c8a571c12", #16
                          "https://www.data.gouv.fr/fr/datasets/r/5b9c29ea-366c-414b-9144-631030293155", #16
                          "https://www.data.gouv.fr/fr/datasets/r/daf5843a-7d1e-4b41-bf8e-220fddffa7ed", #17
                          "https://www.data.gouv.fr/fr/datasets/r/a70a98ba-3462-4f70-8bd4-0e46188c95b6", #17
                          "https://www.data.gouv.fr/fr/datasets/r/43d95da3-997f-4766-b7e4-411d39288364", #18
                          "https://www.data.gouv.fr/fr/datasets/r/14bfa028-44f9-44dc-a1e2-3d355562d8b0", #18
                          "https://www.data.gouv.fr/fr/datasets/r/b870e655-5386-49d4-ba48-920ceb3765f2", #19
                          "https://www.data.gouv.fr/fr/datasets/r/a1e066b7-4bc6-490c-932f-ba3880ab07bf", #19
                          "https://www.data.gouv.fr/fr/datasets/r/24151d9c-1961-4edb-aff9-d3105867f40e", #20X
                          "https://www.data.gouv.fr/fr/datasets/r/b4ff70c3-8374-4b5d-8a31-9dc9cb048c15", #20X
                          "https://www.data.gouv.fr/fr/datasets/r/4b31e0a6-39d6-445c-a283-16d05b078064", #21X
                          "https://www.data.gouv.fr/fr/datasets/r/1c0872a1-3942-49e3-aa6a-1b67f82b2432", #21
                          "https://www.data.gouv.fr/fr/datasets/r/ecda48c2-f09e-4509-951d-e16abe42a872", #22
                          "https://www.data.gouv.fr/fr/datasets/r/9b1987ee-5602-4d70-962b-8c44bb1322b3", #22X
                          "https://www.data.gouv.fr/fr/datasets/r/6b837356-73a6-4e76-9923-d5f91e2ca122", #23
                          "https://www.data.gouv.fr/fr/datasets/r/4cbc2ca6-720f-48d8-ad2c-5c8b3bcc6672", #23
                          "https://www.data.gouv.fr/fr/datasets/r/011f61cc-6789-44ed-bc9a-dad67c1c1039", #24
                          "https://www.data.gouv.fr/fr/datasets/r/ea076ac6-ca6d-4e7d-a632-e11468e84d0d", #24
                          "https://www.data.gouv.fr/fr/datasets/r/929aa82d-3592-4db3-ad90-e1adce979af7", #25X
                          "https://www.data.gouv.fr/fr/datasets/r/384b4e5c-62e9-4e23-89a7-4229aa76f7a8", #25
                          "https://www.data.gouv.fr/fr/datasets/r/bbbc093d-f889-4eb4-9db3-6b5c67cbc563", #26
                          "https://www.data.gouv.fr/fr/datasets/r/f61721dd-7a9e-4194-8afb-2fed18ed03ed", #26
                          "https://www.data.gouv.fr/fr/datasets/r/f4cb9bb9-bb80-41b1-b240-187889d164b8", #27
                          "https://www.data.gouv.fr/fr/datasets/r/1ecdaaae-2f18-4665-b672-c8e773772ea7", #27
                          "https://www.data.gouv.fr/fr/datasets/r/59bc2716-1007-4c2b-bed1-0ba79fc64404", #28
                          "https://www.data.gouv.fr/fr/datasets/r/2ec375f5-5edf-472e-b19c-e62d0e960baf", #28
                          "https://www.data.gouv.fr/fr/datasets/r/8349df0f-a964-4ddb-b4b2-9b8ea09ad970", #29
                          "https://www.data.gouv.fr/fr/datasets/r/6216d652-b35f-4cf9-bff0-0c3dfb28f4a1", #29
                          "https://www.data.gouv.fr/fr/datasets/r/55264b89-1075-497e-8597-9c63422e6f66", #30
                          "https://www.data.gouv.fr/fr/datasets/r/233f6e21-17ae-43e0-8493-1be47c38dc1a", #30
                          "https://www.data.gouv.fr/fr/datasets/r/e518d10b-8100-49dd-953b-b057610b7949", #31
                          "https://www.data.gouv.fr/fr/datasets/r/7b32be5f-ddd0-48a5-ba5b-5c948e38143c", #31
                          "https://www.data.gouv.fr/fr/datasets/r/7cd382a6-f8e9-42d3-85ba-97088e028ffe", #32
                          "https://www.data.gouv.fr/fr/datasets/r/c589ce37-a9a1-4c04-8568-a00f79953a53", #32
                          "https://www.data.gouv.fr/fr/datasets/r/b0e78f3d-9085-4d6d-a47d-6d942f5e9a54", #33
                          "https://www.data.gouv.fr/fr/datasets/r/5f196a76-ba4f-4aa7-af28-eff6c8797b50", #33
                          "https://www.data.gouv.fr/fr/datasets/r/8ad3f9d4-5324-4c74-ae2f-e8b8a89684af", #34
                          "https://www.data.gouv.fr/fr/datasets/r/bedfeab3-55b1-4b2a-964d-4f93b291a101", #34
                          "https://www.data.gouv.fr/fr/datasets/r/8051582c-2f1b-41c6-85a5-2f7f98389193", #35
                          "https://www.data.gouv.fr/fr/datasets/r/42760f87-0231-49b7-ac02-182ccced1f05", #35
                          "https://www.data.gouv.fr/fr/datasets/r/04679829-40f1-4bd2-bb82-23ad526cd95d", #36
                          "https://www.data.gouv.fr/fr/datasets/r/fbe6e16d-1173-4e9f-91fb-2fc15848d508", #36
                          "https://www.data.gouv.fr/fr/datasets/r/62d5d1eb-1ef2-43a5-9bdf-e0b8a67cc6ad", #37
                          "https://www.data.gouv.fr/fr/datasets/r/a5e11f98-cf2b-4169-9d8e-cbc91195dcd0", #37
                          "https://www.data.gouv.fr/fr/datasets/r/e2b0446d-ce6a-4340-b42b-4f33ed227c30", #38
                          "https://www.data.gouv.fr/fr/datasets/r/39f5c4ec-771c-42ed-88e9-49fe20b8878f", #38
                          "https://www.data.gouv.fr/fr/datasets/r/62ee582e-7a3f-43f7-bb1f-eb470ae90d2f", #39
                          "https://www.data.gouv.fr/fr/datasets/r/a40871e3-d4f6-41df-a531-68750361ae31", #39
                          "https://www.data.gouv.fr/fr/datasets/r/8b2d56fc-40df-44b2-9281-7794017d341c", #40
                          "https://www.data.gouv.fr/fr/datasets/r/69b2987e-4216-4708-beb6-7a34d371124b", #40
                          "https://www.data.gouv.fr/fr/datasets/r/c20e431a-98e9-4626-b46a-95a8a9d2d64a", #41
                          "https://www.data.gouv.fr/fr/datasets/r/555f5c1f-e483-4040-9c9a-dc48e4e126a5", #41
                          "https://www.data.gouv.fr/fr/datasets/r/1ed16fc8-31a5-4ef5-b8ca-d32d5b3f4224", #42
                          "https://www.data.gouv.fr/fr/datasets/r/b8ad7e0e-6114-4091-b696-61b9ba203104", #42
                          "https://www.data.gouv.fr/fr/datasets/r/8a4c6cd9-ca20-4913-b541-544f44d73218", #43
                          "https://www.data.gouv.fr/fr/datasets/r/934ab69d-223b-4a77-8962-8c0492763f2b", #43
                          "https://www.data.gouv.fr/fr/datasets/r/6038690c-e555-491f-8588-b595a08c8f67", #44
                          "https://www.data.gouv.fr/fr/datasets/r/92bb91dc-26ca-479e-9476-0b19ae060e30", #44
                          "https://www.data.gouv.fr/fr/datasets/r/09f30cdc-603a-4992-a36a-c36d9201072c", #45
                          "https://www.data.gouv.fr/fr/datasets/r/e2fd770f-55a2-431c-8826-f74128b67e79", #45
                          "https://www.data.gouv.fr/fr/datasets/r/00786c56-23ce-4c05-b91a-97e1faeefb76", #46
                          "https://www.data.gouv.fr/fr/datasets/r/8860f12e-14e9-41bf-8a10-bcb534e0c19e", #46
                          "https://www.data.gouv.fr/fr/datasets/r/7e53e7ec-f71a-43ce-ab93-d31d4f8d0f94", #47
                          "https://www.data.gouv.fr/fr/datasets/r/6fdc9a6e-2577-4205-b9a6-79650cb135d1", #47
                          "https://www.data.gouv.fr/fr/datasets/r/3ab96225-990e-4059-beef-bdf8a9047e12", #48
                          "https://www.data.gouv.fr/fr/datasets/r/1fbbf03e-681e-4d70-adde-aa7107959dd6", #48
                          "https://www.data.gouv.fr/fr/datasets/r/999fe6f2-8de2-4a6f-ae83-d454ac2400d3", #49
                          "https://www.data.gouv.fr/fr/datasets/r/bd96818e-4722-4981-a688-ce6ca64476c6", #49
                          "https://www.data.gouv.fr/fr/datasets/r/582886d0-6f6b-427f-a480-5e82c29de1bd", #50
                          "https://www.data.gouv.fr/fr/datasets/r/77146ae6-b4d4-4694-a9a4-a6b6a3ab72bb", #50
                          "https://www.data.gouv.fr/fr/datasets/r/cd325707-73e3-4f43-b7ad-a7107003b96f", #51
                          "https://www.data.gouv.fr/fr/datasets/r/0c00979a-f90a-4119-b6c5-6be1e72f409a", #51
                          "https://www.data.gouv.fr/fr/datasets/r/224aac45-1e67-4278-9b66-c1a92a0fc765", #52
                          "https://www.data.gouv.fr/fr/datasets/r/8eef7a6c-7ca1-46cf-8b0a-79bc66485cf4", #52
                          "https://www.data.gouv.fr/fr/datasets/r/a5d089b3-746f-4a2c-b116-58d484cad6c3", #53
                          "https://www.data.gouv.fr/fr/datasets/r/6eaaa962-af59-41f0-9151-bc55d69b5910", #53
                          "https://www.data.gouv.fr/fr/datasets/r/34aa794a-ff21-47aa-8f27-7d5ea36ed80d", #54
                          "https://www.data.gouv.fr/fr/datasets/r/ca2f1865-6ab5-4d1e-bfdd-6a17f6efa436", #54
                          "https://www.data.gouv.fr/fr/datasets/r/5c7a4b87-aa26-4c88-9f03-f6986532ca91", #55
                          "https://www.data.gouv.fr/fr/datasets/r/11d12a25-b484-451a-822d-9b6fe05a675f", #55
                          "https://www.data.gouv.fr/fr/datasets/r/570b8bec-5e45-4276-8738-bc0aa82f782a", #56
                          "https://www.data.gouv.fr/fr/datasets/r/74d2f82d-8f4f-4737-9c49-89645223d0cc", #56
                          "https://www.data.gouv.fr/fr/datasets/r/d73f912b-3ac1-4daa-9e3f-587afa6f7887", #57
                          "https://www.data.gouv.fr/fr/datasets/r/a235d996-73d9-4e00-a1ba-31a5cc0c0f33", #57
                          "https://www.data.gouv.fr/fr/datasets/r/83060285-3dbf-40d5-81ea-f873ecb971e1", #58
                          "https://www.data.gouv.fr/fr/datasets/r/f39801a9-d903-4949-bea8-963fc26271f9", #58
                          "https://www.data.gouv.fr/fr/datasets/r/c2a9f75c-0554-48d6-ba1a-534b00063f89", #59
                          "https://www.data.gouv.fr/fr/datasets/r/e356abd8-90ff-4c7f-a0cd-55661c3be402", #59
                          "https://www.data.gouv.fr/fr/datasets/r/8a9591a0-a4b1-4e3e-8610-85540c18d643", #60
                          "https://www.data.gouv.fr/fr/datasets/r/f50d3786-5f1d-4e63-87be-f99222790597", #60
                          "https://www.data.gouv.fr/fr/datasets/r/f07defc4-3281-4104-a468-07807fe43126", #61
                          "https://www.data.gouv.fr/fr/datasets/r/d4992317-c3d2-47f1-a4d7-77bec2532103", #61
                          "https://www.data.gouv.fr/fr/datasets/r/aec69332-1f0a-470f-818a-660d8ade34aa", #62
                          "https://www.data.gouv.fr/fr/datasets/r/b7e514ec-43ee-4536-b7e7-940df1694f24", #62
                          "https://www.data.gouv.fr/fr/datasets/r/4e176848-0138-47c4-ac26-98333d938326", #63
                          "https://www.data.gouv.fr/fr/datasets/r/a881fcb5-e3e2-4193-9e01-7d1879288fba", #63
                          "https://www.data.gouv.fr/fr/datasets/r/8f62ed7c-1d50-4812-9a1d-b549e9f92a33", #64
                          "https://www.data.gouv.fr/fr/datasets/r/9334a953-0e3e-497e-83c2-47228546b4e4", #64
                          "https://www.data.gouv.fr/fr/datasets/r/e7a5a7dd-a1f2-47e6-afed-01133b915051", #65
                          "https://www.data.gouv.fr/fr/datasets/r/b92d64da-48ea-4a8d-9927-590568b3e277", #65
                          "https://www.data.gouv.fr/fr/datasets/r/a095e27f-81eb-4dfa-bfc4-ad01dd0390cd", #66
                          "https://www.data.gouv.fr/fr/datasets/r/a6274453-2e90-42c0-babc-756e9ca7abcb", #66
                          "https://www.data.gouv.fr/fr/datasets/r/ea9a0c19-5519-470e-bb5e-ede4b3301268", #67
                          "https://www.data.gouv.fr/fr/datasets/r/9016ee58-ec7d-4d69-8c9d-3eee8ce78e10", #67
                          "https://www.data.gouv.fr/fr/datasets/r/3d49c539-436f-4893-b5c7-d857fc5edca7", #68
                          "https://www.data.gouv.fr/fr/datasets/r/8383ca46-e2d7-443e-92a5-3fb927168de2", #68
                          "https://www.data.gouv.fr/fr/datasets/r/8dc3b3c0-4edd-4d28-bdd0-4eb2a85aa0d5", #69
                          "https://www.data.gouv.fr/fr/datasets/r/e2243f62-f5d4-4485-81c1-931e61f7505f", #69
                          "https://www.data.gouv.fr/fr/datasets/r/21e70347-f89c-4b45-a409-61b4add6c365", #70
                          "https://www.data.gouv.fr/fr/datasets/r/7937e3dc-18fa-428a-9860-86c7c16745f0", #70
                          "https://www.data.gouv.fr/fr/datasets/r/6101e0eb-2ef5-400f-9b18-87adf71ac572", #71
                          "https://www.data.gouv.fr/fr/datasets/r/fe52e876-259b-4f09-a78f-e134cc8149f2", #71
                          "https://www.data.gouv.fr/fr/datasets/r/98e625c2-58f3-4767-88a6-16fce78c4d4b", #72
                          "https://www.data.gouv.fr/fr/datasets/r/ea958f8c-18dd-4371-8684-2b315a769ff6", #72
                          "https://www.data.gouv.fr/fr/datasets/r/828ec510-e062-48e4-9370-0bd9a5f65a9c", #73
                          "https://www.data.gouv.fr/fr/datasets/r/6ebd3d3a-8753-4378-990d-fa98e977ea59", #73
                          "https://www.data.gouv.fr/fr/datasets/r/23a368d7-59e3-488d-afb7-3890f972dcd7", #74
                          "https://www.data.gouv.fr/fr/datasets/r/212e4fb5-e4fd-4565-af91-1fd340b23a6c", #74
                          "https://www.data.gouv.fr/fr/datasets/r/27bf7b0f-62a8-438f-acf4-b5f58d293322", #75
                          "https://www.data.gouv.fr/fr/datasets/r/aba837dc-fc7c-4010-ab5e-0eb02feb0010", #75
                          "https://www.data.gouv.fr/fr/datasets/r/83047d20-9b28-4734-9aca-1a44801353ee", #76
                          "https://www.data.gouv.fr/fr/datasets/r/273b34d6-88fa-4ebb-b30d-155e69cfcc35", #76
                          "https://www.data.gouv.fr/fr/datasets/r/36a85491-dfb8-45e3-96fa-561158ee34dd", #77
                          "https://www.data.gouv.fr/fr/datasets/r/473c287c-001a-4bfa-8c6c-f59afe61bead", #77
                          "https://www.data.gouv.fr/fr/datasets/r/a7c4b4a1-6b99-46ec-8ca1-b2e368e63a59", #78
                          "https://www.data.gouv.fr/fr/datasets/r/c5aabcc5-3f3d-4caa-ac8a-c766685cf7b5", #78
                          "https://www.data.gouv.fr/fr/datasets/r/0f019547-c446-4d08-930f-9970cfa9b97b", #79
                          "https://www.data.gouv.fr/fr/datasets/r/5bb97008-9c31-4581-acee-6988099ec7a6", #79
                          "https://www.data.gouv.fr/fr/datasets/r/59dd29dd-42c1-480a-83a4-cedc80082816", #80
                          "https://www.data.gouv.fr/fr/datasets/r/15db6936-b20e-4391-a1db-4425dc7e3f25", #80
                          "https://www.data.gouv.fr/fr/datasets/r/c2679caa-f430-4d59-8451-696dd43f8e80", #81
                          "https://www.data.gouv.fr/fr/datasets/r/eef8a8d7-0984-4a75-9a44-834076c1778d", #81
                          "https://www.data.gouv.fr/fr/datasets/r/0d30a20b-7448-4524-8507-6f5765d37b71", #82
                          "https://www.data.gouv.fr/fr/datasets/r/a1c2f5a8-2af7-4874-b4da-7c607358295b", #82
                          "https://www.data.gouv.fr/fr/datasets/r/ffe62e95-57f0-4135-9e28-e95d2e31dede", #83
                          "https://www.data.gouv.fr/fr/datasets/r/5889ea7c-e285-46cb-8ea6-1c317fa4dada", #83
                          "https://www.data.gouv.fr/fr/datasets/r/25e1be54-9384-4632-a5f2-db8f463d3d65", #84
                          "https://www.data.gouv.fr/fr/datasets/r/48ab3fa1-690f-427c-95b2-6ce561dfa5e2", #84
                          "https://www.data.gouv.fr/fr/datasets/r/d5d3514d-867b-4529-a031-1f5e840f816a", #85
                          "https://www.data.gouv.fr/fr/datasets/r/1b35a324-0d51-478c-8d6e-e9f8b4fef8c2", #85
                          "https://www.data.gouv.fr/fr/datasets/r/b69b1bf4-ae45-4061-a10b-8f2a5ed6bafd", #86
                          "https://www.data.gouv.fr/fr/datasets/r/b4e246c6-50d7-43b0-8a0f-a842a37e9f71", #86
                          "https://www.data.gouv.fr/fr/datasets/r/29a456ec-2ce9-4ce1-ba23-4ec3e8b25324", #87
                          "https://www.data.gouv.fr/fr/datasets/r/e7c25f1a-e0bc-4c9b-ba3c-1dca6b6ad8a2", #87
                          "https://www.data.gouv.fr/fr/datasets/r/4343f20d-401d-40cd-82cc-5528fdb66c0d", #88
                          "https://www.data.gouv.fr/fr/datasets/r/f2da33f7-9834-4247-86ba-d661aaf6cfa9", #88
                          "https://www.data.gouv.fr/fr/datasets/r/40c47eaf-09fa-40e3-8a28-0aa5fcc3aa05", #89
                          "https://www.data.gouv.fr/fr/datasets/r/0b3979dc-773d-4888-b94f-e93c5b3634bc", #89
                          "https://www.data.gouv.fr/fr/datasets/r/6969896a-c507-47eb-a3b8-cc8424abd3b3", #90
                          "https://www.data.gouv.fr/fr/datasets/r/4f3ce032-ecdc-43d8-b3dd-3cc42457957b", #90
                          "https://www.data.gouv.fr/fr/datasets/r/04a75839-25fd-46b6-af09-e1934388cef7", #91
                          "https://www.data.gouv.fr/fr/datasets/r/e3719036-f756-4a42-b1d8-ba65235d60e1", #91
                          "https://www.data.gouv.fr/fr/datasets/r/8fc12852-f9b9-47fd-8052-d58cc5ee6a63", #92
                          "https://www.data.gouv.fr/fr/datasets/r/881525f4-f786-4eb6-ac4d-d62f3089ed19", #92
                          "https://www.data.gouv.fr/fr/datasets/r/77927e93-9aa1-462c-bacb-bfff81c8ed48", #93
                          "https://www.data.gouv.fr/fr/datasets/r/8818da6a-3c3a-4c1e-ba7e-1e1a9c66120a", #93
                          "https://www.data.gouv.fr/fr/datasets/r/f302db14-f490-4f73-adf7-a8458ecf1da3", #94
                          "https://www.data.gouv.fr/fr/datasets/r/dc4b8ca5-e72b-4f24-82c6-9d75f08c594b", #94
                          "https://www.data.gouv.fr/fr/datasets/r/c65b7105-6824-4f2e-b53a-66934fbe863b", #95
                          "https://www.data.gouv.fr/fr/datasets/r/00b37d57-76cf-4531-8fde-112a576a9a6e", #95
                          ]
FRANCE_50_25_filepaths_aux = ["https://www.data.gouv.fr/fr/datasets/r/5c857920-ea0d-4b43-ac24-9a7b6beb8051",#1
                              "https://www.data.gouv.fr/fr/datasets/r/98852d18-6580-47d4-9442-0bb3184896c8",#1
                              "https://www.data.gouv.fr/fr/datasets/r/7016326f-e29d-452b-98f8-25a85d65e665",#2
                              "https://www.data.gouv.fr/fr/datasets/r/5aeb7c2c-ecc8-4bec-bbfe-92d60fe6e1eb",#2
                              "https://www.data.gouv.fr/fr/datasets/r/1aaa0a75-6f7f-4937-b99e-92034cb7aa93",#3
                              "https://www.data.gouv.fr/fr/datasets/r/b7368b36-0f4d-4ec7-860b-b17ce032cc29",#3
                              "https://www.data.gouv.fr/fr/datasets/r/d40542f8-cb3e-47b9-a3e9-e4c8d6f7456f",#4
                              "https://www.data.gouv.fr/fr/datasets/r/5b920952-bc3e-4cba-877f-33498f40f48a",#4
                              "https://www.data.gouv.fr/fr/datasets/r/a9f26030-9e53-4bba-ace0-bb4556caf5e7",#5
                              "https://www.data.gouv.fr/fr/datasets/r/b3ac198a-2d45-462e-8491-1f8dfd38c048",#5
                              "https://www.data.gouv.fr/fr/datasets/r/02c4af67-c26d-4389-8ee6-9438ff36ac2b",#6
                              "https://www.data.gouv.fr/fr/datasets/r/473bfd3f-d02e-4182-a6ee-896b863a1447",#6
                              "https://www.data.gouv.fr/fr/datasets/r/473bfd3f-d02e-4182-a6ee-896b863a1447",#7
                              "https://www.data.gouv.fr/fr/datasets/r/51f138f1-f2ff-4947-b384-7eef66186d48",#7
                              "https://www.data.gouv.fr/fr/datasets/r/79440a8b-a62f-4510-8ea7-ad23c662b366",#8
                              "https://www.data.gouv.fr/fr/datasets/r/0b18db3b-0643-46f4-828a-88ac4d72660a",#8
                              "https://www.data.gouv.fr/fr/datasets/r/eab39818-0c34-4da2-bd6e-1dea5c21a081",#9
                              "https://www.data.gouv.fr/fr/datasets/r/09303d58-09e6-4f3d-81eb-5d5eb8ca63bb",#9
                              "https://www.data.gouv.fr/fr/datasets/r/e70d42d7-5f95-49cb-921b-e8ed8983437c",#10
                              "https://www.data.gouv.fr/fr/datasets/r/d37beffd-513e-48e8-a81a-17752acffa2d",#10
                              "https://www.data.gouv.fr/fr/datasets/r/766b194d-2462-4674-8837-65babfd053ad",#11
                              "https://www.data.gouv.fr/fr/datasets/r/2458bcb0-c613-42d4-999b-a3f5095548b8",#11
                              "https://www.data.gouv.fr/fr/datasets/r/3f44d999-4c66-4d29-a6ff-bcd102820483",#12
                              "https://www.data.gouv.fr/fr/datasets/r/969b0825-f378-4b4a-b440-f01293e29f1e",#12
                              "https://www.data.gouv.fr/fr/datasets/r/dac8dcb5-f5fd-4b33-b8b2-b4933b5e2960",#13
                              "https://www.data.gouv.fr/fr/datasets/r/be21ebc9-68f5-4da1-b428-864dfedf0d28",#13
                              "https://www.data.gouv.fr/fr/datasets/r/8f2178fe-5541-4cd4-9625-39562ac9301b",#14
                              "https://www.data.gouv.fr/fr/datasets/r/5b53ba30-16a4-476c-8591-d0af01c657e8",#14
                              "https://www.data.gouv.fr/fr/datasets/r/4830d2b7-3ead-46cd-af18-ac271401c24c",#15
                              "https://www.data.gouv.fr/fr/datasets/r/6fa02577-ed6d-4054-b1cf-229cba678bb3",#15
                              "https://www.data.gouv.fr/fr/datasets/r/0dcb8222-3b9d-4b97-b6ec-fa0ec6dcb81f",#16
                              "https://www.data.gouv.fr/fr/datasets/r/63de734f-b208-43cb-8687-491a6a1f0f4c",#16
                              "https://www.data.gouv.fr/fr/datasets/r/564f08d8-2721-4815-b365-91f7b3bd7be4",#17
                              "https://www.data.gouv.fr/fr/datasets/r/71ec691d-6903-473f-ae66-b1e388ce5994",#17
                              "https://www.data.gouv.fr/fr/datasets/r/04262393-70ab-480f-aaf1-222b32608abf",#18
                              "https://www.data.gouv.fr/fr/datasets/r/555bddb8-e31b-4884-ad97-41dbe79909fa",#18
                              "https://www.data.gouv.fr/fr/datasets/r/577b28eb-97d1-4fde-8d7c-dec676134243",#19
                              "https://www.data.gouv.fr/fr/datasets/r/e0d14030-8396-4d79-a373-14744f90c57a",#19
                              "https://www.data.gouv.fr/fr/datasets/r/0213bd75-2c2a-4990-84f9-8bdb7fd4cf45",#20
                              "https://www.data.gouv.fr/fr/datasets/r/01903148-ef18-49ff-92a3-d65f13ad7118",#20
                              "https://www.data.gouv.fr/fr/datasets/r/5dbd84c4-23dc-47aa-ac50-d77d9da8bb07",#21
                              "https://www.data.gouv.fr/fr/datasets/r/43b6f534-3e94-4e70-9cbd-479ac64e90a9",#21
                              "https://www.data.gouv.fr/fr/datasets/r/dcfa050f-d597-4079-abee-16a1709ed0b7",#22
                              "https://www.data.gouv.fr/fr/datasets/r/41db67c6-0305-47d0-b468-b756d55880ff",#22
                              "https://www.data.gouv.fr/fr/datasets/r/c1411629-1b8e-458c-b8a0-47025ed07d0b",#23
                              "https://www.data.gouv.fr/fr/datasets/r/21584f02-1805-47a2-94f7-51e0e9edb36e",#23
                              "https://www.data.gouv.fr/fr/datasets/r/5a097fd2-866a-4e19-9f6f-2515f99dc4a5",#24
                              "https://www.data.gouv.fr/fr/datasets/r/f34b6baa-713a-474e-9c8d-fc19a3673bbb",#24
                              "https://www.data.gouv.fr/fr/datasets/r/db182768-26da-4d1e-bf04-e8c2a6952129",#25
                              "https://www.data.gouv.fr/fr/datasets/r/386d1feb-c5a0-449a-99b9-acc722bc7428",#25
                              "https://www.data.gouv.fr/fr/datasets/r/8e5c00b3-4109-46c6-b5ff-22dfe0f3d796",#26
                              "https://www.data.gouv.fr/fr/datasets/r/1dccf9c7-ffa6-4d7a-b02b-545fff50db42",#26
                              "https://www.data.gouv.fr/fr/datasets/r/9dc66a12-8c72-4145-aa88-870ddcd9ef29",#27
                              "https://www.data.gouv.fr/fr/datasets/r/282ffe66-f317-4a84-af33-9e5192ceec49",#27
                              "https://www.data.gouv.fr/fr/datasets/r/b5f23bdd-3923-41cb-9c22-8bbbf11b4558",#28
                              "https://www.data.gouv.fr/fr/datasets/r/d4a4d677-0945-4bcd-a012-2603c1e5cdcf",#28
                              "https://www.data.gouv.fr/fr/datasets/r/4f01bc95-b3a5-4ff1-92e9-9305b69ecdfa",#29
                              "https://www.data.gouv.fr/fr/datasets/r/f7fd3e97-fccf-4b96-b739-d82cfa547e74",#29
                              "https://www.data.gouv.fr/fr/datasets/r/9cf6422d-fc8e-4538-b0cb-45ee3eda6eec",#30
                              "https://www.data.gouv.fr/fr/datasets/r/708475c2-31b7-4019-adac-a6cb252fdd71",#30
                              "https://www.data.gouv.fr/fr/datasets/r/9d1466ea-9d3d-4e57-91b3-c622392a1ca6",#31
                              "https://www.data.gouv.fr/fr/datasets/r/ed8dff23-2a5d-4e28-91af-5e5370608832",#31
                              "https://www.data.gouv.fr/fr/datasets/r/16e36d68-a165-46cf-bbd1-807d84b902bc",#32
                              "https://www.data.gouv.fr/fr/datasets/r/bad57106-30f1-471b-8cda-312f6c5c7da9",#32
                              "https://www.data.gouv.fr/fr/datasets/r/73e35d27-f8ed-4671-89bc-aee31edaa265",#33
                              "https://www.data.gouv.fr/fr/datasets/r/fcf1a1be-d999-46d1-b2a6-f7ef4193ba50",#33
                              "https://www.data.gouv.fr/fr/datasets/r/1afecd10-f2ab-46f9-abf3-a1c60a535eeb",#34
                              "https://www.data.gouv.fr/fr/datasets/r/1afc2021-3f24-4285-90c1-4589b08b8a3e",#34
                              "https://www.data.gouv.fr/fr/datasets/r/5706dd09-5646-438a-b3dc-6f90a8acdff0",#35
                              "https://www.data.gouv.fr/fr/datasets/r/45ffe00f-6b32-4093-93b3-66c0eef3704e",#35
                              "https://www.data.gouv.fr/fr/datasets/r/ec242a9d-de2d-44cb-93e1-38e9e446f906",#36
                              "https://www.data.gouv.fr/fr/datasets/r/8357423c-2066-4069-8df6-6d3f1c0882e4",#36
                              "https://www.data.gouv.fr/fr/datasets/r/e2713080-a3fc-4263-91fe-89462e23934b",#37
                              "https://www.data.gouv.fr/fr/datasets/r/507bebc5-e3f8-4e38-8013-0a8eb7e0c912",#37
                              "https://www.data.gouv.fr/fr/datasets/r/a3f04096-7338-405b-8ba4-cdceb27c308f",#38
                              "https://www.data.gouv.fr/fr/datasets/r/50851f3b-6e64-4100-98a7-7938ca467ac9",#38
                              "https://www.data.gouv.fr/fr/datasets/r/d587bf9c-c03b-4472-a762-cca4bb73d918",#39
                              "https://www.data.gouv.fr/fr/datasets/r/0c057bc0-1f5a-4693-9236-4753194312de",#39
                              "https://www.data.gouv.fr/fr/datasets/r/ee18c83d-6abd-4912-bc4b-f438826afaab",#40
                              "https://www.data.gouv.fr/fr/datasets/r/c20a5528-aeed-43f9-9c64-abac5c18ac99",#40
                              "https://www.data.gouv.fr/fr/datasets/r/5b76eca2-5205-4c4c-8970-b09b87a43fa3",#41
                              "https://www.data.gouv.fr/fr/datasets/r/1d5af114-d127-4916-ad54-2033e92ec733",#41
                              "https://www.data.gouv.fr/fr/datasets/r/2c575634-22cf-487a-b6b3-d23855d5ca54",#42
                              "https://www.data.gouv.fr/fr/datasets/r/9d971455-807c-4941-bb81-075672d2972a",#42
                              "https://www.data.gouv.fr/fr/datasets/r/b5351be9-c84a-4275-a036-36128af5c2b6",#43
                              "https://www.data.gouv.fr/fr/datasets/r/ebdc6f17-6ebe-4ef3-a8ea-fe97f70a71db",#43
                              "https://www.data.gouv.fr/fr/datasets/r/a4df4b5a-32c5-44be-906c-3431ca2daddb",#44
                              "https://www.data.gouv.fr/fr/datasets/r/ab7c3eeb-f2bd-4b6c-94cc-c2c075d6fd61",#44
                              "https://www.data.gouv.fr/fr/datasets/r/b49cd816-49a7-4a35-8b9d-bb688999fb1f",#45
                              "https://www.data.gouv.fr/fr/datasets/r/3290dc78-b793-49b8-9696-c26f8dbd0358",#45
                              "https://www.data.gouv.fr/fr/datasets/r/9f51dfcd-e775-4dd5-9d2e-de7f10417472",#46
                              "https://www.data.gouv.fr/fr/datasets/r/cf698aeb-dd64-4737-b391-a1c03b94abf1",#46
                              "https://www.data.gouv.fr/fr/datasets/r/c7be854a-48f7-4cb3-81b0-e56164d1100d",#47
                              "https://www.data.gouv.fr/fr/datasets/r/dbbabdca-44b9-4808-9681-25d9234865f1",#47
                              "https://www.data.gouv.fr/fr/datasets/r/74afcdaa-4dac-40ad-b03f-8671654700a7",#48
                              "https://www.data.gouv.fr/fr/datasets/r/85b6f125-c890-4977-b80e-1bfccdd9599f",#48
                              "https://www.data.gouv.fr/fr/datasets/r/b80dc79e-02e4-4d87-842f-d893c26b46bf",#49
                              "https://www.data.gouv.fr/fr/datasets/r/dec500c7-eb18-4396-9386-bcde950b5994",#49
                              "https://www.data.gouv.fr/fr/datasets/r/a2f56896-1c49-4ec9-9373-8b55e2e56412",#50
                              "https://www.data.gouv.fr/fr/datasets/r/acf4e65b-a3b4-4e72-a273-3d96962065c3",#50
                              "https://www.data.gouv.fr/fr/datasets/r/c8f364da-f475-40ff-8670-ac0311b4d0e5",#51
                              "https://www.data.gouv.fr/fr/datasets/r/6e6b6f55-ef02-406a-a91c-6dc592d0b81c",#51
                              "https://www.data.gouv.fr/fr/datasets/r/0f7e7fea-b935-4767-a428-c77fdb33c695",#52
                              "https://www.data.gouv.fr/fr/datasets/r/4c9f7e99-0497-45ec-9815-9c4341657959",#52
                              "https://www.data.gouv.fr/fr/datasets/r/068a319c-6eb0-4f11-ba6f-665063dc302d",#53
                              "https://www.data.gouv.fr/fr/datasets/r/1a35d638-11e9-41f9-9849-ea40f596118c",#53
                              "https://www.data.gouv.fr/fr/datasets/r/f8545b54-b582-4629-88b8-842008adbbd9",#54
                              "https://www.data.gouv.fr/fr/datasets/r/d6c59c1f-1edb-40a1-a55d-442dbc77f277",#54
                              "https://www.data.gouv.fr/fr/datasets/r/653dcc95-9407-4028-ba63-169f22f01851",#55
                              "https://www.data.gouv.fr/fr/datasets/r/355b3f63-0228-4adf-b282-afe2ae93d8a2",#55
                              "https://www.data.gouv.fr/fr/datasets/r/b88fcb14-5517-491b-be8f-fab848659f3a",#56
                              "https://www.data.gouv.fr/fr/datasets/r/3d5dd19c-fb73-400a-9c0b-63825cd345a8",#56
                              "https://www.data.gouv.fr/fr/datasets/r/c95204f9-808b-45f4-be77-5635c6dff434",#57
                              "https://www.data.gouv.fr/fr/datasets/r/be39a594-b424-4885-a228-89fa66763444",#57
                              "https://www.data.gouv.fr/fr/datasets/r/bb794929-8586-4f73-be74-c4c17178a5f3",#58
                              "https://www.data.gouv.fr/fr/datasets/r/01d32911-5e37-4ed1-820f-10532c4909c1",#58
                              "https://www.data.gouv.fr/fr/datasets/r/33d7a588-5c61-41a7-b0f0-8e0895419d9b",#59
                              "https://www.data.gouv.fr/fr/datasets/r/1e2097b8-7b60-4bab-820e-7b593823ab52",#59
                              "https://www.data.gouv.fr/fr/datasets/r/00562ea8-de39-4aff-8cbf-2b5bc17b1278",#60
                              "https://www.data.gouv.fr/fr/datasets/r/420f2480-f524-4666-a65e-65bc6a316a20",#60
                              "https://www.data.gouv.fr/fr/datasets/r/4fb3e309-b7ed-4181-b217-ae9cdbae1f25",#61
                              "https://www.data.gouv.fr/fr/datasets/r/d8408da4-ff13-4332-a1a3-51c97a318acd",#61
                              "https://www.data.gouv.fr/fr/datasets/r/0ed9e61b-7da0-47b9-a8dd-1e94595f12b0",#62
                              "https://www.data.gouv.fr/fr/datasets/r/c5245a05-9f90-412f-9ef3-651c345fa6f3",#62
                              "https://www.data.gouv.fr/fr/datasets/r/dd8aabaa-db0b-4c3c-b3e2-ccb377ad5f31",#63
                              "https://www.data.gouv.fr/fr/datasets/r/6ef48ae4-63d7-4f0b-b07e-2a2c6778c171",#63
                              "https://www.data.gouv.fr/fr/datasets/r/a4a522d5-8d86-49cc-90ae-607e2719ee72",#64
                              "https://www.data.gouv.fr/fr/datasets/r/12e92e16-1905-485e-9fe1-7c8089179fb9",#64
                              "https://www.data.gouv.fr/fr/datasets/r/44eac5d0-c984-4912-9c81-ad148258cc4c",#65
                              "https://www.data.gouv.fr/fr/datasets/r/c72e96f9-9011-45de-96b0-7d1450089b8b",#65
                              "https://www.data.gouv.fr/fr/datasets/r/76a7eeb3-1d80-43df-a345-8fabbc1350cf",#66
                              "https://www.data.gouv.fr/fr/datasets/r/ba7cbc38-17ed-4f00-9e2e-461183695345",#66
                              "https://www.data.gouv.fr/fr/datasets/r/0b4ff3cc-a1af-45d4-af1a-aa2047bd7a08",#67
                              "https://www.data.gouv.fr/fr/datasets/r/43a46b30-5ddf-414e-834c-8197158b9101",#67
                              "https://www.data.gouv.fr/fr/datasets/r/95685984-4d8d-427f-887d-a31c9f6023a5",#68
                              "https://www.data.gouv.fr/fr/datasets/r/ad78221e-0a05-4052-94a4-93f7b0c487c7",#68
                              "https://www.data.gouv.fr/fr/datasets/r/50a0cde3-0663-40a3-baab-2f3f57489362",#69
                              "https://www.data.gouv.fr/fr/datasets/r/e42439db-6684-4ac3-9ea4-e9748ffa1fac",#69
                              "https://www.data.gouv.fr/fr/datasets/r/e9b52f15-9386-4e19-8c53-8f4cc4f7fe70",#70
                              "https://www.data.gouv.fr/fr/datasets/r/de4b1a05-d7e7-477a-af82-d2178d5f69cf",#70
                              "https://www.data.gouv.fr/fr/datasets/r/e7f7454d-8b18-49b0-94d6-cb0c8c37adbe",#71
                              "https://www.data.gouv.fr/fr/datasets/r/1aee8bc6-e80f-486b-b213-e8fb7b6b0a43",#71
                              "https://www.data.gouv.fr/fr/datasets/r/c590a664-8c3b-471d-8b78-d666dbde8de5",#72
                              "https://www.data.gouv.fr/fr/datasets/r/ba177ba8-6b78-453e-a7d0-4602fcc0bb1e",#72
                              "https://www.data.gouv.fr/fr/datasets/r/94757a8e-fa4a-4608-b846-316502983c4a",#73
                              "https://www.data.gouv.fr/fr/datasets/r/ed0878f5-5935-4a38-9cc2-abcad94a9d52",#73
                              "https://www.data.gouv.fr/fr/datasets/r/9096f970-f711-4db5-9f22-77618cb50729",#74
                              "https://www.data.gouv.fr/fr/datasets/r/2fd676a6-c8e2-47df-adb1-974a60a5482b",#74
                              "https://www.data.gouv.fr/fr/datasets/r/e318ff1d-afab-494d-ab45-263bdc5cc2e2",#75
                              "https://www.data.gouv.fr/fr/datasets/r/863521b9-9e04-4255-b720-6d3ff4da3ef1",#75
                              "https://www.data.gouv.fr/fr/datasets/r/00ca5488-9af6-443f-9ff5-50f311db1951",#76
                              "https://www.data.gouv.fr/fr/datasets/r/b8a0151c-9e65-4dda-874b-c389ad0d9513",#76
                              "https://www.data.gouv.fr/fr/datasets/r/b1c1f113-5eb6-4d96-acc9-a6cd9d2bbf92",#77
                              "https://www.data.gouv.fr/fr/datasets/r/2b592ebd-d8c4-4fbd-8205-e04fab626ca8",#77
                              "https://www.data.gouv.fr/fr/datasets/r/4bae389f-573a-4b0b-8f44-70685378ac7a",#78
                              "https://www.data.gouv.fr/fr/datasets/r/587bcad8-7514-45fa-a023-39ddd5215433",#78
                              "https://www.data.gouv.fr/fr/datasets/r/eef615a3-6f60-48b4-8bff-ad4ebb56f6fc",#79
                              "https://www.data.gouv.fr/fr/datasets/r/df43b49b-8aba-417b-8668-5de5f9959fdf",#79
                              "https://www.data.gouv.fr/fr/datasets/r/776efd91-dd9d-408e-82f5-eef083647a41",#80
                              "https://www.data.gouv.fr/fr/datasets/r/71ba2168-b4ce-4857-b531-2af1237f7bf7",#80
                              "https://www.data.gouv.fr/fr/datasets/r/a1c59a03-fcf3-4bb4-ba8b-0548d8340af1",#81
                              "https://www.data.gouv.fr/fr/datasets/r/addb5f72-5c42-4536-b298-e0be6eeeebbb",#81
                              "https://www.data.gouv.fr/fr/datasets/r/3e6512ed-c624-48cf-a181-fabba4fddbee",#82
                              "https://www.data.gouv.fr/fr/datasets/r/b3c484e1-3a71-4f59-940e-e67e94c661f8",#82
                              "https://www.data.gouv.fr/fr/datasets/r/b940cc00-a9b7-442d-82ef-cbb57155b445",#83
                              "https://www.data.gouv.fr/fr/datasets/r/795008f3-ea36-4764-81c4-a32d115203c0",#83
                              "https://www.data.gouv.fr/fr/datasets/r/0ba3a902-039b-48e3-9d81-dffca71a43c7",#84
                              "https://www.data.gouv.fr/fr/datasets/r/9fa267b6-b3cf-4e37-83c1-93f7081b431a",#84
                              "https://www.data.gouv.fr/fr/datasets/r/f32d0d52-e3ae-4a77-b5c2-309a1ead12c4",#85
                              "https://www.data.gouv.fr/fr/datasets/r/581556e3-017a-48d2-9da1-7f9475e028f6",#85
                              "https://www.data.gouv.fr/fr/datasets/r/e9df0d95-437a-43b3-b904-b9a407e062e5",#86
                              "https://www.data.gouv.fr/fr/datasets/r/fa9bcd39-7354-4ed3-961f-b779e59ef1f4",#86
                              "https://www.data.gouv.fr/fr/datasets/r/1a208da5-d58f-4dac-af86-c1a99aaebaa6",#87
                              "https://www.data.gouv.fr/fr/datasets/r/0bce6ef5-dcba-4d40-803c-e943ed92c88e",#87
                              "https://www.data.gouv.fr/fr/datasets/r/4e804296-d0e0-47f8-ad0d-7b5871e11af5",#88
                              "https://www.data.gouv.fr/fr/datasets/r/6dba459b-4697-4543-a83d-d6feafe79bbe",#88
                              "https://www.data.gouv.fr/fr/datasets/r/e26211d0-598c-41d4-a342-ec059b9df660",#89
                              "https://www.data.gouv.fr/fr/datasets/r/4bf9f60b-769c-423d-a50b-a37880e50776",#89
                              "https://www.data.gouv.fr/fr/datasets/r/19676267-5cdf-4897-8c42-6ac53e3dc8e9",#90
                              "https://www.data.gouv.fr/fr/datasets/r/08385fd6-7db5-4eb3-8501-e7023c79c323",#90
                              "https://www.data.gouv.fr/fr/datasets/r/113208b4-6f16-46c3-8503-54f1f12e13e2",#91
                              "https://www.data.gouv.fr/fr/datasets/r/0bce2def-d015-41e4-9d14-5127eed7b94d",#91
                              "https://www.data.gouv.fr/fr/datasets/r/4563a4bd-cec7-431c-bafc-13dba4b4412c",#92
                              "https://www.data.gouv.fr/fr/datasets/r/634cc1ec-7d2a-47e4-9ef1-be51a894dbba",#92
                              "https://www.data.gouv.fr/fr/datasets/r/abd67960-6c7c-43b8-9a3a-4eafd95ace97",#93
                              "https://www.data.gouv.fr/fr/datasets/r/6eadda53-ef58-4246-91ac-e84ed1a75e8f",#93
                              "https://www.data.gouv.fr/fr/datasets/r/6688191d-33d1-478a-93c6-a8eab9b784d9",#94
                              "https://www.data.gouv.fr/fr/datasets/r/4ff5e1a0-c949-47ce-bdbf-e625c6df3f0e",#94
                              "https://www.data.gouv.fr/fr/datasets/r/00a8e54f-5ce9-4882-b959-a08edbf649ca",#95
                              "https://www.data.gouv.fr/fr/datasets/r/b3482517-0713-4eac-afcc-eb5b135a6ff2"] #95]

"""
This function takes each url that downloads a csv and reads the contents into a pandas dataframe. All data that is
beyond the minimum date is dropped and the duplicate date column is dropped.
"""
def read_and_parse_csv_filtered(url, min_date='2014-01-01'):
    min_date = pd.to_datetime(min_date)
    df = pd.read_csv(url, compression='gzip', delimiter=';')
    df['Date'] = pd.to_datetime(df['AAAAMMJJ'].astype(str), format='%Y%m%d', errors='coerce')
    df = df[df['Date'] >= min_date].copy()
    df['Date'] = df['Date'].dt.strftime('%Y%m%d')
    df.drop(columns=['AAAAMMJJ'], inplace=True)

    return df

"""
This function joins the two weather datasets into one dataframe. The unique identifiers for the two
datasets are the NUM_POSTE and Date columns. For all duplicated columns in the datasets, the auxillary
dataset is less validated, so a tag is added to these columns for future cleaning or removal. 
"""
def process_file_pair(main_url, aux_url, min_date='2014-01-01'):
    main_df = read_and_parse_csv_filtered(main_url, min_date)
    aux_df = read_and_parse_csv_filtered(aux_url, min_date)

    # Only proceed if both dataframes are not empty
    if not main_df.empty and not aux_df.empty:
        joined_df = pd.merge(main_df, aux_df, on=['NUM_POSTE', 'Date'], how='inner', suffixes=('', '_aux'))
        return joined_df
    
    else:
        return pd.DataFrame()

"""
    This is the helper function to the above merging function. It unites the two filepath lists
    adds them to a larger dataset if they return any data when read or united. The larger dataframe
    is created and a department code column is added to allow for joining each department to their
    region.
"""
def join_all_file_pairs(main_urls, aux_urls, min_date='2014-01-01'):
    file_pairs = list(zip(main_urls, aux_urls))
    joined_chunks = []
    # Process each pair sequentially
    for main_url, aux_url in file_pairs:
        joined_df = process_file_pair(main_url, aux_url, min_date)
        if not joined_df.empty:
            joined_chunks.append(joined_df)

    if joined_chunks:
        combined = pd.concat(joined_chunks, ignore_index=True)
        combined["departement_code"] = combined["NUM_POSTE"] // 1000000
        return combined
    else:
        return pd.DataFrame()


FRANCE_energy_meteo_daily = join_all_file_pairs(FRANCE_50_25_filepaths, FRANCE_50_25_filepaths_aux)

"""
This section of code reads the department and region mapping dataset to merge onto the weather dataframe
to allow for regional aggregations. Corse's A and B labels are removed and duplicates are removed to
expite join calculations. Joins are on the department code, creating a dataset at the weather station per day level.
"""
dep_regions = pd.read_csv(region_depts_filepath)
dep_regions['num_dep'] = dep_regions['num_dep'].str.replace('[AB]', '0', regex=True).astype(int)
dep_regions = dep_regions.drop_duplicates(subset='num_dep')

FRANCE_energy_meteo_daily = pd.merge(FRANCE_energy_meteo_daily, dep_regions, left_on="departement_code", right_on="num_dep", how="left")

# These features represent the meaningful measurements in the two weather datasets. A commented out
# feature indicates it has a very high covarriance with another feature, so was removed in feature 
# selection. 23 features make the final dataframe. 

aggregations = {
    'RR': 'mean',
    'ALTI': 'mean',
    #'TN': 'mean',
    'PMERM' : 'mean',
    'INST' : 'mean',
    'GLOT' : 'mean',
#    'UM' : 'mean',
    'DHUMI40' : 'mean',
#    'DHUMI80' : 'mean',
#    'TSVM' : 'mean',
#    'ETPMON' : 'mean',
    'NEIGETOTX' : 'mean',
    'NEIG' : 'mean',
    'BROU' : 'mean',
    'ORAG' : 'mean',
    'GRESIL' : 'mean',
    'GRELE' : 'mean',
    'ROSEE' : 'mean',
    'VERGLAS' : 'mean',
    'SOLNEIGE' : 'mean',
    'GELEE' : 'mean',
    'FUMEE' : 'mean',
    'BRUME' : 'mean',
    'ECLAIR' : 'mean',
    'TX': 'mean',
    #'TM': 'mean',
    #'TAMPLI' : 'mean',
    'TNSOL' : 'mean',
    #"TN50" : 'mean',
    'FXY' : 'mean',
    #'FXI' : 'mean',
    #'FXI3S' : 'mean',
    #'DG' : 'mean',
    #'FFM' : 'mean',
    #'FF2M' : 'mean',
    #'FXI2' : 'mean',
    'DRR': 'mean'
}
# The aggregtion of each feature brings the dataset to a daily measure by region. 
FRANCE_energy_meteo_daily = FRANCE_energy_meteo_daily.groupby(["Date", 'region_name']).agg(aggregations).reset_index()
# This new year column allows for future joins on population
FRANCE_energy_meteo_daily['year'] = FRANCE_energy_meteo_daily['Date'].astype(str).str[:4]

"""
This section of code reads the population per region dataset into a dataframe, removing unnecessary 
columns like area and unique id. Since the yearly population is a column, a pivot, or melt, is 
performed to make the years a seperate column, with a second populatuion column. This will join
correctly to the regional weather data. After the join on region and year, the duplicate columns are 
dropped.
"""
dep_pop = pd.read_excel(population_filepath) 
dep_pop.drop(columns=['objectid'], inplace=True)
pop_long = dep_pop.melt(
    id_vars=['nom_reg', 'superf'],  
    value_vars=[col for col in dep_pop.columns if col.startswith('p')],
    var_name='pop_year',
    value_name='population'
)
pop_long['year'] = '20' + pop_long['pop_year'].str[1:3]
pop_long["density"] = pop_long["population"] / pop_long["superf"]

FRANCE_energy_meteo_daily = pd.merge(FRANCE_energy_meteo_daily, pop_long, left_on=['region_name', 'year'],right_on=['nom_reg', 'year'], how='left')
FRANCE_energy_meteo_daily.drop(columns=['nom_reg', 'year', 'pop_year'], inplace=True)

"""
This section of code reads the consumption dataset and all metrics besides electricity consumption per region.
The dataset is at the half-hour interval per region. All NAs are dropped for the target
feature, with further data cleaning. The data is aggregated to the day and region level and joined
on the weather data. 
"""
cons_df = pd.read_csv(consumption_filepath, delimiter=';', decimal=",")

cons_df.drop(columns=["Date - Heure",
        "flag_ignore",
        "Consommation brute gaz (MW PCS 0C) - NaTran", # GRTgaz  renamed to NaTran
        "Statut - NaTran",
        "Consommation brute gaz (MW PCS 0C) - Terga",
        "Statut - Terga",
        "Consommation brute gaz totale (MW PCS 0C)",
        "Consommation brute totale (MW)"
    ],
    inplace=True
)

cons_df.rename(columns={"Consommation brute lectricit (MW) - RTE":"Consumption"}, inplace=True)
cons_df = cons_df.dropna(subset=['Consumption'])
cons_df['Date'] = pd.to_datetime(cons_df['Date'], errors='coerce') 
cons_df['Date'] = cons_df['Date'].dt.strftime('%Y%m%d')
cons_df = cons_df.groupby(["Date", 'Rgion']).agg({'Consumption': 'sum'}).reset_index()
FRANCE_energy_meteo_daily = pd.merge(FRANCE_energy_meteo_daily, cons_df, left_on=['region_name', 'Date'],right_on=['Rgion', 'Date'], how='left')

"""
This section of code reads the holiday schedule dataset and only keeps the two needed columns to join
onto the larger dataset. The day of weather measurements adds a new column for weekday
to check if it is a typical working day or not. The holiday data set is joined on
the weather data, and if its not a holiday and not a weekend, the day is considered
a working day. After this calculation, the holiday and weekend columns are dropped.
"""
holiday_df = pd.read_csv(holiday_filepath)
holiday_df["Holiday?"] = True
holiday_df = holiday_df[['date', 'Holiday?']]
holiday_df.rename(columns={'date':'Date'}, inplace=True)
holiday_df['Date'] = pd.to_datetime(holiday_df['Date'], errors='coerce')  # ensure it's datetime
FRANCE_energy_meteo_daily['Date'] = pd.to_datetime(FRANCE_energy_meteo_daily['Date'], errors='coerce')  # ensure it's datetime
FRANCE_energy_meteo_daily['Weekend?'] = FRANCE_energy_meteo_daily['Date'].dt.dayofweek >= 5
FRANCE_energy_meteo_daily = pd.merge(FRANCE_energy_meteo_daily, holiday_df, on='Date', how='left')
FRANCE_energy_meteo_daily['Workday?'] = (~(FRANCE_energy_meteo_daily["Holiday?"] | FRANCE_energy_meteo_daily['Weekend?'])).astype(int)
FRANCE_energy_meteo_daily.drop(columns=['Holiday?','Weekend?'], inplace=True)

# After all data is joined and cleaned, the dataset is saved locally. 
FRANCE_energy_meteo_daily.to_csv('france_energy_meteo_daily.csv', index=False)
```
